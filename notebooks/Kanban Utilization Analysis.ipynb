# Kanban Process Model
* $\lambda$ = New work arrival rate
* $\mu$ = Work completion rate

![kanban](kanban.svg)



\begin{equation*}
\rho  = \frac{\lambda}{\mu} \\
N(M/M/1)  = \frac{\rho}{1 -  \rho}
\end{equation*}

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
from IPython.display import Image
%matplotlib notebook
```

```python
rtd = nx.DiGraph()

rtd.add_nodes_from(list(range(0,6)))

forward_edges=[(x,x+1,{'label':'λ'}) for x in range(0,5)]
rtd.add_edges_from(forward_edges)

reverse_edges=[(6-x,6-x-1,{'label':'µ'}) for x in range(1,6)]
rtd.add_edges_from(reverse_edges)
rtd.add_edge(5,'...',label='λ')
rtd.add_edge('...',5,label='µ')

nx.nx_agraph.write_dot(rtd,'multi.dot')
dg = nx.nx_agraph.to_agraph(rtd)
dg.layout(prog='dot',args="-Grankdir=LR")
dg.draw('multi.png', format='png',prog='dot')
Image("multi.png")
```

# M/M/1 Rate Transition Diagram
![Rate State Diagram](multi.png)

```python

pBusy = np.arange(0,1,0.01)
pIdle = 1 - pBusy
fig=plt.subplot()
fig.set_xlabel('p')
fig.set_ylabel('N')
fig.vlines(.8, 0,100,colors='r')
fig.plot(pBusy, pBusy/pIdle)
```
